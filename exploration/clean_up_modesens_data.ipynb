### Clean up Color

```python
# ! pip install webcolors
# import numpy as np
import webcolors
def extract_colors(text):
    valid_colors = []
    if isinstance(text, str):
        text = text.strip("['").strip("']").lower()
        colors = text.split(',')
        for c in colors:
            for cc in c.split('/'):
                cc = cc.strip()
                cc = cc.strip("'")
                if cc in webcolors.CSS3_NAMES_TO_HEX:
                    valid_colors.append(cc)
    return valid_colors
```

```python
products["colors"] = products["color"].apply(extract_colors)
products.head()
```

### Extract categories

```python
category_keywords = pd.read_csv('./data/category_keywords.csv', index_col=False)
category_keywords['category'] = category_keywords['category'].apply(lambda x: x.lower())
category_keywords.head()
```

```python
products['categories'] = products['description'].apply(lambda x: [keyword for keyword in category_keywords['category'].unique() if isinstance(x, str) and keyword in x])
```

### Extract texture/fabric

```python
products['textures'] = products['description'].apply(lambda x: [keyword for keyword in texture_keywords['texture'].unique() if isinstance(x, str) and keyword in x])
```

### Keyword extraction

```python
# read fashion keywords
fashion_keywords = []
with open('./data/fashion_keywords_cleaned.txt') as f:
    for line in f:
        fashion_keywords.append(line.strip())
```

```python
# check if fashion keywords are in the description
products["description"] = products["description"].apply(lambda x: " ".join([keyword for keyword in fashion_keywords if keyword in x]))
products.head()
```

```python
# drop rows with no named categories and textures and colors
products = products[products['categories'].apply(lambda x: len(x) > 0) & products['textures'].apply(lambda x: len(x) > 0) & products['colors'].apply(lambda x: len(x) > 0)]
products
```

```python
def new_description(row):
    colors = ' '.join(row['colors'])
    textures = ' '.join(row['textures'])
    if row.main_category == 'clothing':
        categories = ''.join(row['categories'])
    elif row.main_category == 'accessories':
        categories = row['categories'][0]
    else:
        categories = row.sub_category
    return colors + ' ' + textures + ' ' + categories
```

```python
products['new_description'] = products.apply(new_description, axis=1)
products
```

```python
# drop _id, color, textures, categories
products.drop(columns=['_id', 'color', 'textures', 'categories'], inplace=True)
products.to_csv('./data/modesens_products_sample_cleaned.csv',index=False)
```
