```python
import faiss
import numpy as np

# Generate random data
title_embed = sample['title_embed'].apply(lambda x:np.array(x[1:-1].split(',')).astype('float32'))
title_embed = np.concatenate(title_embed).astype('float32').reshape(len(title_embed),-1)

# Create an index: IndexFlatL2 measures the L2 (or Euclidean) distance between all given points between our query vector, and the vectors loaded into the index. Itâ€™s simple, very accurate, but not too fast.
title_index = faiss.IndexFlatL2(title_embed.shape[1])

# Add data to the index
title_index.add(title_embed)


# Generate random data
descrip_embed = sample['descrip_embed'].apply(lambda x:np.array(x[1:-1].split(',')).astype('float32'))
descrip_embed = np.concatenate(descrip_embed).astype('float32').reshape(len(descrip_embed),-1)

# Create an index: IndexFlatL2 measures the L2 (or Euclidean) distance between all given points between our query vector, and the vectors loaded into the index. Itâ€™s simple, very accurate, but not too fast.
descrip_index = faiss.IndexFlatL2(descrip_embed.shape[1])

# Add data to the index
descrip_index.add(descrip_embed)
```

```python
from matplotlib import pyplot as plt
from PIL import Image
import requests

# Perform a similarity search
k = 5
item = items[1]
query_vector = model.encode(item).reshape(1,-1)

_, descrip_indices = descrip_index.search(query_vector, k)
_, title_indices = title_index.search(query_vector, k)

fig = plt.figure(figsize=(10, 7))

# setting values to rows and column variables
print('chatgpt says: ',item)
print('description suggests: ')

for i in range(k):
    # Adds a subplot at the 1st position
    
    fig.add_subplot(1, k, i+1)
    try:
        url = sample.loc[descrip_indices[0][i],'image_link']
        im = Image.open(requests.get(url,stream=True,timeout=(10, 20)).raw)
  
        # showing image
        plt.imshow(im)
        plt.axis('off')
        plt.title(descrip_indices[0][i])

    except:
        print('link expired: ',i,sample.loc[descrip_indices[0][i],'description'])
plt.show()

print('title suggests: ')

fig = plt.figure(figsize=(10, 7))
for i in range(k):
    # Adds a subplot at the 1st position
    
    fig.add_subplot(1, k, i+1)
    try:
        url = sample.loc[title_indices[0][i],'image_link']
        im = Image.open(requests.get(url,stream=True,timeout=(10, 20)).raw)
  
        # showing image
        plt.imshow(im)
        plt.axis('off')
        plt.title(title_indices[0][i])

    except:
        print('link expired: ',i,sample.loc[title_indices[0][i],'title'])
plt.show()
```

```python
from matplotlib import pyplot as plt
from PIL import Image
import requests

def get_recommendation(item,k):

    query_vector = model.encode(item).reshape(1,-1)

    _, descrip_indices = descrip_index.search(query_vector, k)
    _, title_indices = title_index.search(query_vector, k)

    fig = plt.figure(figsize=(10, 7))

    # setting values to rows and column variables
    print('chatgpt says: ',item)
    print('description suggests: ')

    for i in range(k):
        # Adds a subplot at the 1st position
        
        fig.add_subplot(1, k, i+1)
        try:
            url = sample.loc[descrip_indices[0][i],'image_link']
            im = Image.open(requests.get(url,stream=True,timeout=(10, 20)).raw)
    
            # showing image
            plt.imshow(im)
            plt.axis('off')
            plt.title(descrip_indices[0][i])

        except:
            print('link expired: ',i,sample.loc[descrip_indices[0][i],'description'])
    plt.show()

    print('title suggests: ')

    fig = plt.figure(figsize=(10, 7))
    for i in range(k):
        # Adds a subplot at the 1st position
        
        fig.add_subplot(1, k, i+1)
        try:
            url = sample.loc[title_indices[0][i],'image_link']
            im = Image.open(requests.get(url,stream=True,timeout=(10, 20)).raw)
    
            # showing image
            plt.imshow(im)
            plt.axis('off')
            plt.title(title_indices[0][i])

        except:
            print('link expired: ',i,sample.loc[title_indices[0][i],'title'])
    plt.show()

    return descrip_indices, title_indices
```

```python
k = 5  # Number of nearest neighbors to retrieve
descrip_results = []
title_results = []

for item in items:
    descrip_indices,title_indices = get_recommendation(item,k)
    descrip_results.append(descrip_indices)
    title_results.append(title_indices)
```
